# Heart Failure Survival Prediction Project

## Project Overview
This project focuses on predicting the survival outcomes of heart failure patients using clinical and demographic attributes. Heart disease is one of the leading causes of death worldwide, and understanding the factors that influence survival rates is critical for early intervention and improved patient outcomes.

The goal of this project is to apply **machine learning techniques** to build and compare predictive models that can forecast whether a heart failure patient will survive or not. In addition to building accurate models, this project aims to identify the key clinical and demographic features that contribute most to survival predictions. 

## Dataset Information
The dataset used for this project comes from the **[UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/519/heart+failure+clinical+records)**. It contains 299 records of patients who experienced heart failure, with 13 clinical and demographic features, along with the target variable `DEATH_EVENT`, which indicates whether the patient survived (0) or died (1).

### Key Features
- **Age**: Age of the patient
- **Anaemia**: Decrease of red blood cells or hemoglobin (Boolean)
- **Creatinine Phosphokinase (CPK)**: Level of CPK enzyme in blood (mcg/L)
- **Diabetes**: Whether the patient has diabetes (Boolean)
- **Ejection Fraction**: Percentage of blood leaving the heart at each contraction
- **High Blood Pressure**: Whether the patient has hypertension (Boolean)
- **Platelets**: Platelet count (kiloplatelets/mL)
- **Serum Creatinine**: Level of serum creatinine (mg/dL)
- **Serum Sodium**: Level of serum sodium (mEq/L)
- **Sex**: Male (1) or Female (0)
- **Smoking**: Whether the patient is a smoker (Boolean)
- **Time**: Follow-up period (days)

### Target Variable
- **DEATH_EVENT**: 
    - 1 = Patient died
    - 0 = Patient survived

## Feature Dictionary

| Feature | Description |
|---|---|
| **age** | Patient's age in years. Older patients are typically at higher risk. |
| **anaemia** | Whether the patient has low red blood cells (1 = Yes, 0 = No). Anaemia can worsen heart failure outcomes. |
| **creatinine_phosphokinase** | Level of CPK enzyme in the blood (mcg/L). High levels may indicate heart muscle damage. |
| **diabetes** | Whether the patient has diabetes (1 = Yes, 0 = No). Diabetes increases heart failure risk. |
| **ejection_fraction** | Percentage of blood leaving the heart at each contraction. Low values indicate poor heart function. |
| **high_blood_pressure** | Whether the patient has hypertension (1 = Yes, 0 = No). High BP adds stress to the heart. |
| **platelets** | Platelet count (kiloplatelets/mL). May indicate blood clotting ability or potential abnormalities. |
| **serum_creatinine** | Level of creatinine in the blood (mg/dL). High levels indicate potential kidney problems. |
| **serum_sodium** | Level of sodium in the blood (mEq/L). Low levels indicate fluid retention and severe heart failure. |
| **sex** | Patient's gender (1 = Male, 0 = Female). Gender can influence survival rates. |
| **smoking** | Whether the patient smokes (1 = Yes, 0 = No). Smoking worsens heart and blood vessel health. |
| **time** | Follow-up period (days). Shorter times may indicate early death. |
| **DEATH_EVENT** | Target variable (1 = Died, 0 = Survived). |

## Data Preprocessing and Feature Engineering

In this section, we perform essential **data preprocessing** steps to prepare the dataset for modeling. This includes:

- **Log Transformation**: Right-skewed numeric features, such as `serum_creatinine`, `creatinine_phosphokinase` (CPK), and `platelets`, are log-transformed to reduce skewness and handle extreme values more effectively.
- **Feature Engineering**: New features are created to enhance model interpretability and performance.
    - `age_group`: Patients are grouped into meaningful age categories (e.g., 40-49, 50-59) to capture non-linear relationships between age and survival.
    - Additional interaction terms or combined risk scores may be considered based on initial exploratory analysis.
- **Standardization**: All **numeric features** are standardized (mean=0, std=1) to ensure consistent scaling across variables. This is particularly important for distance-based models (SVM, KNN) and gradient-based optimization algorithms (logistic regression, neural networks).

These preprocessing steps aim to improve model performance by addressing **outliers, skewed distributions, feature scales, and potential non-linear relationships**.


### Log Transformations

Certain features in this dataset, such as **serum creatinine**, **creatinine phosphokinase (CPK)**, and **platelet count**, exhibit **right-skewed distributions with long tails**. 

To **reduce skewness** and **lessen the impact of extreme outliers**, a **log transformation** is applied to these features. This transformation compresses the scale, ensuring that **extremely high values do not dominate the model training process**, while preserving the relative ranking of values.

The `np.log1p()` function is used instead of `np.log()` to safely handle any **zero values** (although in this dataset, zero values are unlikely for these features).


```python
# Log transformations
df['log_creatinine'] = np.log1p(df['serum_creatinine'])
df['log_cpk'] = np.log1p(df['creatinine_phosphokinase'])
df['log_platelets'] = np.log1p(df['platelets'])
```

### Age Ranges (Binning)

Age is a **continuous variable**, but in clinical practice, patients are often grouped into **age ranges or categories** to better capture **age-related risk patterns**. 

By creating **age ranges**, we can:
- Simplify the interpretation of models.
- Capture potential **non-linear relationships** between age and survival, where risk may sharply increase in older age groups.
- Align the analysis with **standard clinical age groupings**, making results easier to communicate to healthcare professionals.

This step helps the model **learn differences between younger, middle-aged, and older patients**, rather than assuming the relationship between age and survival is strictly linear.

```python
df['age_group'] = pd.cut(df['age'], 
                         bins=[40, 49, 59, 69, 79, 89, 100], 
                         labels=['40-49', '50-59', '60-69', '70-79', '80-89', '90+'])
```

### Encoding Age Group

Since `age_group` is an **ordered categorical feature**, there are two possible encoding techniques:

- **One-Hot Encoding** treats each age group as an independent category.
- **Ordinal Encoding** assigns increasing numeric values to older age groups, preserving the natural order.

Both approaches are created to allow flexibility during modeling. The final choice of encoding will depend on which method provides better model performance.


```python
df = pd.get_dummies(df, columns=['age_group'], drop_first=True, dtype=int)  #One-Hot Encoding
```

### Feature Engineering: Clinical Categories

To enhance interpretability and capture **non-linear relationships**, two continuous variables (`ejection_fraction` and `time`) are **binned into clinically meaningful categories**.

- **Ejection Fraction (EF)** is grouped into:

```python
df['ef_category'] = pd.cut(df['ejection_fraction'], 
                           bins=[0, 30, 50, 100], 
                           labels=['Severe', 'Moderate', 'Normal'])
```

### Handling Imbalanced Classes

The `DEATH_EVENT` variable is imbalanced, with **67% survivors** and **33% deaths**. Imbalanced data can bias models like Logistic Regression to favor the majority class (survived), leading to poor recall for the minority class (died).

To address this imbalance, we use the **`class_weight='balanced'`** option in Logistic Regression. This automatically assigns **higher weights to the minority class** (deaths) and **lower weights to the majority class** (survived), helping the model focus more on correctly predicting deaths.

This technique helps improve:
- **Recall for the minority class.**
- **Overall model balance between precision and recall.**

The impact of this balanced weighting will be compared to models trained without class balancing to evaluate its effectiveness.

### Logistic Regression

```python
def train_logistic_regression(X_train, X_test, y_train, y_test, feature_set, param_grid):
    print(f"Training Logistic Regression with feature set: {feature_set}\n")
    
    # Subset the features
    X_train = X_train[feature_set]
    X_test = X_test[feature_set]
    
    # Logistic Regression with Grid Search
    lr = LogisticRegression(solver='liblinear', max_iter=1000, class_weight='balanced')
    grid = GridSearchCV(lr, param_grid, cv=5, scoring='f1')
    grid.fit(X_train, y_train)

    # Best Model
    best_model = grid.best_estimator_
    print(f"Best Parameters: {grid.best_params_}\n")

    # Predictions & Evaluation
    y_pred = best_model.predict(X_test)
    y_prob = best_model.predict_proba(X_test)[:, 1]

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    print(f"ROC AUC Score: {roc_auc_score(y_test, y_prob):.4f}\n")

    return best_model
```

```python
def plot_roc_curve(model, X_test, y_test, label):
    """Plots the AUC-ROC curve for a given model."""
    # Get predicted probabilities
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Compute ROC curve and AUC score
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.3f})')

    return roc_auc  # Return AUC score to compare later
```

### Addressing Class Imbalance with SMOTE

The original dataset is **highly imbalanced**, with significantly more survivors than deaths. This class imbalance can bias the model, making it favor the majority class (survived) while underperforming in identifying the minority class (died). To mitigate this, **SMOTE (Synthetic Minority Oversampling Technique)** is applied to the training data. 

SMOTE generates **synthetic examples of the minority class (deaths)**, balancing the class distribution in the training set. This allows the model to learn more effectively about both classes, particularly the underrepresented death class, potentially improving its ability to identify high-risk patients.


```python
def train_and_evaluate_with_smote(X_train, X_test, y_train, y_test, feature_set, feature_set_name):
    print(f"\nTraining Logistic Regression with SMOTE - {feature_set_name}")

    # Subset data to the selected features
    X_train_sub = X_train[feature_set]
    X_test_sub = X_test[feature_set]

    # Apply SMOTE to training data
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train_sub, y_train)

    # Logistic Regression with Grid SearchCV (no need for class_weight because SMOTE balanced the classes)
    lr = LogisticRegression(solver='liblinear', max_iter=1000)
    grid = GridSearchCV(lr, param_grid, cv=5, scoring='f1')
    grid.fit(X_train_smote, y_train_smote)

    # Get best model
    best_model = grid.best_estimator_
    print(f"Best Parameters: {grid.best_params_}")

    # Predict on original imbalanced test set
    y_pred = best_model.predict(X_test_sub)
    y_prob = best_model.predict_proba(X_test_sub)[:, 1]

    # Print results
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    roc_auc = roc_auc_score(y_test, y_prob)
    print(f"ROC AUC Score: {roc_auc:.4f}\n")

    return best_model, roc_auc
```

### K-Nearest Neighbors (KNN)

K-Nearest Neighbors (KNN) is a **distance-based classification algorithm** that predicts the class of a new data point by **examining the labels of the `k` closest points (neighbors)** in the training data. The majority class among these neighbors determines the predicted class.

KNN is a **non-parametric model**, meaning it makes no strong assumptions about the underlying data distribution. This makes it particularly **flexible** and useful for datasets where the relationship between features and the target is complex or non-linear.

However, KNN can be **sensitive to irrelevant features and unscaled data**, so **standardizing numeric features** (which we already did) is critical. Additionally, since the data is imbalanced, **SMOTE-balanced training data will be used** to give each class equal representation during training.

KNN is evaluated across the same **four feature sets** used for logistic regression, allowing for direct comparison between the algorithms.

### Introduction to Random Forest

Random Forest is an **ensemble learning algorithm** that builds multiple decision trees during training and combines their predictions to improve accuracy and robustness. Each tree is trained on a **random subset** of the data, and at each split, only a **random subset of features** is considered. This randomness helps **reduce overfitting** and makes Random Forest a highly **robust and versatile model**.

#### Key Advantages
- Handles **non-linear relationships** and complex interactions between features.
- Provides **feature importance scores**, making it useful for understanding which variables contribute most to predictions.
- Performs well on datasets with a **mix of numerical and categorical variables**.
- Can handle **missing data** and **imbalanced datasets**, especially when combined with techniques like **SMOTE**.

#### Why Its Suitable for This Project
In a clinical setting where patient outcomes depend on a complex combination of **demographic, clinical, and biochemical factors**, Random Forests ability to capture interactions between variables makes it a **strong candidate for predicting heart failure survival**. By applying **SMOTE-balanced training data**, we can further ensure the model gets sufficient exposure to high-risk patients, improving its ability to correctly identify patients at risk of death.

The following section will train and evaluate Random Forest models using the same structured process applied to Logistic Regression, comparing performance across:
- **Original features**
- **Binned features**
- **Log-transformed features**
- **Log-transformed + Binned features**

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200],         # Number of trees
    'max_depth': [5, 10, 15],          # Tree depth (None means fully grown trees)
    'min_samples_split': [2, 5, 10],         # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],           # Minimum samples in a leaf node
    'max_features': ['sqrt', 'log2']         # Number of features to consider at each split
}
```

```python
def train_and_evaluate_random_forest_smote(X_train, X_test, y_train, y_test, feature_set, feature_set_name):
    print(f"\nTraining Random Forest with SMOTE - {feature_set_name}")

    # Subset features
    X_train_sub = X_train[feature_set]
    X_test_sub = X_test[feature_set]

    # Apply SMOTE to balance training data
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train_sub, y_train)

    # Define Random Forest with grid search
    rf = RandomForestClassifier(random_state=42)
    grid = GridSearchCV(rf, param_grid_rf, cv=5, scoring='f1', n_jobs=-1)
    grid.fit(X_train_smote, y_train_smote)

    # Best model
    best_model = grid.best_estimator_
    print(f"Best Parameters: {grid.best_params_}")

    # Predict on original (imbalanced) test set
    y_pred = best_model.predict(X_test_sub)
    y_prob = best_model.predict_proba(X_test_sub)[:, 1]

    # Evaluation
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    roc_auc = roc_auc_score(y_test, y_prob)
    print(f"ROC AUC Score: {roc_auc:.4f}\n")

    return best_model, roc_auc
```

### Introduction to Support Vector Machines (SVM)

Support Vector Machines (SVM) are powerful **supervised learning algorithms** that are particularly effective for classification problems with **clear class boundaries**. SVM works by finding a **hyperplane** that best separates data points from different classes, maximizing the **margin** between the two classes. 

#### Why It's Suitable for This Project
SVM is particularly useful when the **number of features is relatively small compared to the number of samples**, which fits this dataset well after feature engineering. ### Introduction to Decision Trees

Decision Trees are **simple yet powerful classification algorithms** that split data based on feature thresholds to create a **tree-like structure** for decision making. Each internal node represents a decision rule based on a feature, and each leaf node represents a classification outcome.

#### Why Its Suitable for This Project
Decision Trees are particularly good at **handling mixed data types (numerical, categorical)** and can naturally **handle non-linear relationships**. With proper pruning and hyperparameter tuning, they can generalize well even on smaller datasets like this heart failure survival dataset.

```python
param_grid_dt = {
    'max_depth': [3, 5, 10, 15, None],            # Control tree size
    'min_samples_split': [2, 5, 10],               # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],                  # Minimum samples per leaf
    'criterion': ['gini', 'entropy']                # Impurity measures
}
```

```python
def train_and_evaluate_decision_tree_smote(X_train, X_test, y_train, y_test, feature_set, feature_set_name):
    print(f"\nTraining Decision Tree with SMOTE - {feature_set_name}")

    # Subset data to selected feature set
    X_train_sub = X_train[feature_set]
    X_test_sub = X_test[feature_set]

    # Apply SMOTE to training set
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train_sub, y_train)

    # Decision Tree with grid search
    dt = DecisionTreeClassifier(random_state=42)
    grid = GridSearchCV(dt, param_grid_dt, cv=5, scoring='f1', n_jobs=-1)
    grid.fit(X_train_smote, y_train_smote)

    # Best model
    best_model = grid.best_estimator_
    print(f"Best Parameters: {grid.best_params_}")

    # Predict on the imbalanced test set
    y_pred = best_model.predict(X_test_sub)
    y_prob = best_model.predict_proba(X_test_sub)[:, 1]

    # Evaluation
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    roc_auc = roc_auc_score(y_test, y_prob)
    print(f"ROC AUC Score: {roc_auc:.4f}\n")

    return best_model, roc_auc
```
